
**The NumPy library **

is one of the foundational libraries for data science and data analysis in Python. It provides powerful tools for working with large, multi-dimensional arrays and matrices, along with a collection of mathematical functions to operate on these arrays efficiently.

Key Features of NumPy


*   Efficient Array Storage: NumPy arrays (ndarrays) are more memory-efficient and provide faster data processing than Python lists.
*   Mathematical Operations: NumPy provides built-in functions for mathematical operations, which are faster than operations on lists.
* Broadcasting: Allows operations on arrays of different shapes, which makes it possible to perform element-wise operations without explicit loops.
*Vectorization: NumPy’s internal implementation allows you to perform operations on entire arrays at once without loops, making computations faster and more concise.

NumPy includes built-in functions for many common operations in data analysis:

* Mean: np.mean(array) calculates the average.

* Median: np.median(array) finds the median value.

* Standard Deviation: np.std(array) measures the spread of the data.
* Matrix Multiplication: np.dot(array1, array2) performs matrix multiplication.
* Reshaping: array.reshape(new_shape) changes the shape of an array without modifying the data.







Example of Calculating Mean and Standard Deviation

```python
import numpy as np
#a list with numbers
regular_list = [1,2,3,4,5]
#creating a numpy array from the list
np_array = np.array(regular_list)
print(np_array)
```

```python
#array with student scores
scores = np.array([70,85,89,78,65])
#calculate statistics
average_score = np.mean(scores)
highest_score = np.max(scores)
score_variance = np.std(scores)

print(average_score)
print(highest_score)
print(score_variance)
```
